## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Reading Loan Train File

```python
loan = pd.read_csv('train_file.csv')
loan.head(25)
```

```python
loan.info()
```

```python
loan.describe()
```

```python
loan = loan.drop('Loan_ID',axis=1)
```

## Dealing With Missing Values 

```python
loan.isnull().sum()
```

```python
loan['Gender'].fillna(loan['Gender'].mode()[0],inplace=True)
```

```python
loan['Dependents'].fillna(loan['Dependents'].mode()[0],inplace=True)
```

```python
loan['Self_Employed'].fillna(loan['Self_Employed'].mode()[0],inplace=True)
```

```python
loan['LoanAmount'].fillna(loan['LoanAmount'].mean(),inplace=True)
```

```python
loan['Loan_Amount_Term'].fillna(loan['Loan_Amount_Term'].median(),inplace=True)
```

```python
loan['Credit_History'].fillna(loan['Credit_History'].median(),inplace=True)
```

```python
loan.dropna(inplace=True)
```

```python
loan.isnull().sum()
```

## Visualize Data

## Convert Categorical Values into Numerical

```python
Gender=pd.get_dummies(loan['Gender'],drop_first=True)
Gender.head(5)
```

```python
Married=pd.get_dummies(loan['Married'],drop_first=True)
Married.head(5)
```

```python
Dependents=pd.get_dummies(loan['Dependents'],drop_first=True)
Dependents.head(5)
```

```python
Education=pd.get_dummies(loan['Education'],drop_first=True)
Education.head(5)
```

```python
Self_Employed=pd.get_dummies(loan['Self_Employed'],drop_first=True)
Self_Employed.head(5)
```

```python
Property_Area=pd.get_dummies(loan['Property_Area'],drop_first=True)
Property_Area.head(5)
```

```python
Loan_Status=pd.get_dummies(loan['Loan_Status'],drop_first=True)
Loan_Status.head(5)
```

```python
loan=pd.concat([loan,Gender,Married,Dependents,Education,Self_Employed,Property_Area,Loan_Status],axis=1)
loan.head(5)
```

```python
loan.drop(['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status'],axis=1,inplace=True)
```

```python
loan.head(10)
```

```python
loan.shape
```

## Train Data

```python
X = loan.drop('Y',axis=1)
Y = loan['Y']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(x_train, y_train)
logistic.coef_
```

```python
logistic.intercept_
```

```python
predict = logistic.predict(x_test)
print(predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix = confusion_matrix(y_test, predict)
```

```python
print(confusion_matrix)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test,predict)
```

```python
print(acc)
```

```python
print("miscclassified saamples: %d" % (y_test!=predict).sum())
```
